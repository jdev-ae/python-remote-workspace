```python
df.dropna(inplace=True)
df.query('price != "Rent on Request"', inplace=True)
```

```python
is_studio = df.bedrooms.str.contains("Room")
is_lac = df.price.str.contains("Lac")
is_cm = df.price.str.contains(",")
```

```python
df.loc[is_studio, 'bedrooms'] = '0 Bedrooms'
```

```python
df.bedrooms = df.bedrooms.str.extract("(\\d+)").astype(float)
```

```python
df.loc[is_lac, 'price'] = df.loc[is_lac, 'price'].str.extract("(\\d+\\.?\\d*)").astype("float") * 1e5
```

```python
del df['bathrooms']
```

```python
df.loc[is_cm, 'price'] = df.loc[is_cm, 'price'].str.replace(',', '').astype("float")
```
