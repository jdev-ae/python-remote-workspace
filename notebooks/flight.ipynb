```python
import pandas as pd
from IPython.display import display_html
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
data = [
    [8, 45, 1, 126],
    [4, 0, 1, 145],
    [8, 20, 1, 145],
    [13, 45, 1, 171],
    [16, 55, 1, 171],
    [11, 0, 1, 171],
    [14, 10, 1, 171],
    [19, 50, 1, 171],
    [4, 35, 1, 176], 
    [4, 20, 1, 192],
    [2, 25, 1, 201],
    [2, 25, 0, 396],
    [3, 55, 1, 322],
    [4, 0, 1, 236],
    [4, 15, 1, 240],
    [4, 20, 1, 327],
    [4, 25, 1, 201],
    [4, 30, 1, 548],
    [4, 40, 1, 304],
    [4, 45, 1, 548],
    [4, 55, 1, 502],
    [17, 22, 1, 999]
]
```

```python
df = pd.DataFrame(data)
```

```python
df.columns = ['hrs', 'mins', 'stops', 'price']
```

```python
def convert_to_min(row):
    row['mins'] += row['hrs'] * 60
    return row
```

```python
df.apply(convert_to_min, axis=1)
del df['hrs']
```

```python
display_side_by_side(df.head(), df.tail())
```

```python
df_zscore = (df - df.mean())/df.std()
```

```python
df_zscore
```

```python
df_zscore = df_zscore.drop(df_zscore[abs(df_zscore.mins) > 2.5].index)
df_zscore = df_zscore.drop(df_zscore[abs(df_zscore.price) > 2.5].index)
```

```python
df_zscore
```

```python
df = df.loc[df_zscore.index]
df.columns = ['mins', 'stops', 'price']
```

```python
df
```

```python
time_df = df.sort_values(by=['mins', 'price', 'stops'])
price_df = df.sort_values(by=['price', 'mins', 'stops'])
```

```python
display_side_by_side(time_df, price_df)
```

```python
t_idx = list(time_df.index.values)
p_idx = list(price_df.index.values)
```

```python
t_idx_diff = [abs(t_idx.index(i) - p_idx.index(i)) for i in t_idx]
```

```python
time_df['diff'] = t_idx_diff
```

```python
time_df
```

```python
final_df = time_df[time_df['diff'] == time_df['diff'].min()].sort_values(by=['price', 'mins', 'stops'])
```

```python
final_df.reset_index(drop=True)
```
