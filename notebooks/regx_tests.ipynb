#### Date format:  `now +/- y m d`
[regex101](https://regex101.com/r/V8aIOD/2)

```python
r1 = r'^now$|^now ?(?P<op>-|\+) ?((?P<years>[1-3])y)? ?((?P<months>[1-9]|1[01])m)? ?((?P<days>[1-9]|[1-5]\d)d)?$'
r1 = re.compile(r1)
```

```python
r1_test1 = 'now - 3y 3m 59d'
```

```python
r1_res1 = r1.search(r1_test1)
```

```python
print(r1_res1)
# If match founds return object else returns None
```

```python
op = r1_res1.group('op')
years = r1_res1.group('years')
months = r1_res1.group('months')
days = r1_res1.group('days')
```

```python
print(f'op: {op}, years {years}, months {months}, days {days}')
```

```python
r1_test2 = 'now + 100y'
r1_res2 = r1.search(r1_test2)
print(r1_res2)
```

#### Check date from year 2000 (`YYYY-MM-DD`)
[regex101](https://regex101.com/r/3GwFR0/1)

```python
r2 = r'^20\d{2}-(?:0\d|1[0-2])-(?:3[01]|[12][0-9]|0[1-9])$'
# no groups in this regx disabled greedy using ?: in groups
# results would be None/some value
r2 = re.compile(r2)
```

```python
r2_test1 = '2019-09-02'
print(r2.search(r2_test1))
```

```python
r2_res1 = r2.search(r2_test1)
```

```python
print(r2_res1.groupdict())
# groupdict is empty 
```

```python
r2_test2 = '1999-09-02'
print(r2.search(r2_test2))
```
