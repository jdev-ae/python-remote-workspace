```python
def correct_price(row):
    price = row['price']
    try:
        row['price'] = float(price)
    except:
        if ',' in price:
            row['price'] = float(price.replace(',', ''))
        elif 'Lac' in price:
            row['price'] = float(price.replace('Lac', '')) * 1e5
    return row
```
